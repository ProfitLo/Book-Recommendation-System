```python
ratings=pd.read_csv('Ratings.csv',sep=",",error_bad_lines=False,encoding='latin-1')
```

```python
ratings.head()
```

```python
ratings.rename(columns={'User-ID':'user_id',"Book-Rating":'book_rating'},inplace=True)
```

```python
x=ratings['user_id'].value_counts()>200
```

```python
x[x].shape
```

```python
y=x[x].index
y
```

```python
ratings=ratings[ratings['user_id'].isin(y)]
```

```python
ratings.shape
```

```python
ratings_with_books=ratings.merge(books,on='ISBN')
```

```python
ratings_with_books
```

```python
ratings_with_books.shape
```

```python
number_rating=ratings_with_books.groupby('title')['book_rating'].count().reset_index()
```

```python
number_rating.rename(columns={'book_rating':'number of rating'},inplace=True)
```

```python
number_rating
```

```python
final_rating=ratings_with_books.merge(number_rating,on='title')
```

```python
final_rating.head()
```

```python
final_rating=final_rating[final_rating['number of rating']>=50]
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot=final_rating.pivot_table(columns='user_id',index='title',values='book_rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distances,suggestion=model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distances
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

```python
distances,suggestion=model.kneighbors(book_pivot.iloc[201,:].values.reshape(1,-1),n_neighbors=6)
```

```python
for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

```python
np.where(book_pivot.index=='Animal Farm')[0][0]
```

```python
def recommend_book(book_name):
    book_id=np.where(book_pivot.index==book_name)[0][0]
    distances,suggestion=model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)
    for i in range(len(suggestion)):
        if i!=0:
            print(book_pivot.index[suggestion[i]])
        else:
            print("The suggestion for",book_name,"are :",book_pivot.index[suggestion[i]])
```

```python
recommend_book('Animal Farm')
```
